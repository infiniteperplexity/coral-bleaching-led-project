```python
def is_black(x, y):
    return img[x, y, 0] != 255

black_regions = []
checked = set()

import sys
sys.setrecursionlimit(1_000_000)
def flood_fill(x = 0, y = 0, id = 0):
    checked.add((x, y))
    if is_black(x, y):
        black_regions[id].append((x, y))
        for dirs in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            new_x, new_y = x + dirs[0], y + dirs[1]
            if (new_x, new_y) not in checked and 0 <= new_x < 1995 and 0 <= new_y < 2000:
                flood_fill(new_x, new_y, id)


for i in range(1995):
    for j in range(2000):
        if (i, j) not in checked and is_black(i, j):
            black_regions.append([])
            flood_fill(i, j, len(black_regions) - 1)

print(len(black_regions))
```
